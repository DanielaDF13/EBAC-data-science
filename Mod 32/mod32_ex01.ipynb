# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import math
import matplotlib.pyplot as plt
```

1. Uma máquina embala ração para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatório, com distribuição Normal, com média 15kg, mas com desvio padrão 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg.

```python
# Parâmetros da distribuição normal
media = 15
desvio_padrao = 0.1
valor = 15.2

# Cálculo da probabilidade
prob = 1 - stats.norm.cdf(valor, loc=media, scale=desvio_padrao)

print(f"a probabilidade de que um pacote selecionado aleatoriamente dessa produção tenha mais que 15,2 kg é de: {(prob)*100:.2f}%")
```

2. Da máquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parâmetros da distribuição normal
media = 15
desvio_padrao = 0.1
q = 0.95

# Cálculo da probabilidade
prob =stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f"O Peso é de {(prob):.2f} Kg com 95% de confiança")
```

3. Suponha duas variáveis aleatórias $Y_1$ e $Y_2$, a primeira com distribuição Normal Padrão (média=0, variância=1), a segunda com distribuição *t-studdent* com 5 graus de liberdade (também padrão). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na célula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu código começa aqui

# Parâmetros da distribuição normal
media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

# Cálculo das probabilidades para a distribuição normal padrão
prob_normal = stats.norm.cdf(Y, loc=media, scale=desvio_padrao)

# Parâmetros da distribuição  t-studdent 
graus_de_liberdade=5

# Cálculo das probabilidades para a distribuição t-student com 5 graus de liberdade
prob_tstudent = stats.t.cdf(Y, df=graus_de_liberdade)
```

4. Faça uma visualização gráfica dos dados do item anterior.

```python
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_normal, label="Distribuição Normal Padrão", color='blue')
plt.plot(Y, prob_tstudent, label="Distribuição t-Student (5 graus de liberdade)", color='red')
plt.title("Probabilidade acumulada para diferentes distribuições")
plt.xlabel("Valor de Y")
plt.ylabel("Probabilidade acumulada")
plt.legend()
plt.grid(True)
plt.show()
```
