# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
import math
import matplotlib.pyplot as plt
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
# Parmetros da distribuio normal
media = 15
desvio_padrao = 0.1
valor = 15.2

# Clculo da probabilidade
prob = 1 - stats.norm.cdf(valor, loc=media, scale=desvio_padrao)

print(f"a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg  de: {(prob)*100:.2f}%")
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Parmetros da distribuio normal
media = 15
desvio_padrao = 0.1
q = 0.95

# Clculo da probabilidade
prob =stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f"O Peso  de {(prob):.2f} Kg com 95% de confiana")
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)
# Seu cdigo comea aqui

# Parmetros da distribuio normal
media = 0
variancia = 1
desvio_padrao = math.sqrt(variancia)

# Clculo das probabilidades para a distribuio normal padro
prob_normal = stats.norm.cdf(Y, loc=media, scale=desvio_padrao)

# Parmetros da distribuio  t-studdent 
graus_de_liberdade=5

# Clculo das probabilidades para a distribuio t-student com 5 graus de liberdade
prob_tstudent = stats.t.cdf(Y, df=graus_de_liberdade)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
plt.figure(figsize=(10, 6))
plt.plot(Y, prob_normal, label="Distribuio Normal Padro", color='blue')
plt.plot(Y, prob_tstudent, label="Distribuio t-Student (5 graus de liberdade)", color='red')
plt.title("Probabilidade acumulada para diferentes distribuies")
plt.xlabel("Valor de Y")
plt.ylabel("Probabilidade acumulada")
plt.legend()
plt.grid(True)
plt.show()
```
